## Webscrapping of IMDB show rating of top 250 shows.

```python
# Importing the required libraries

import openpyxl
from bs4 import BeautifulSoup
import requests
```

```python
# Create new workbook

file=openpyxl.Workbook()
print(file.sheetnames)
sheet = file.active
sheet.title = 'Top Rated Shows'
print(file.sheetnames)
sheet.append(['Rank','Show Name','Year of Release','IMDB Ratings'])
```

```python
# Scrapping the shows in the web page
try:
    source = requests.get('https://www.imdb.com/chart/toptv/')
    source.raise_for_status()

    soup = BeautifulSoup(source.text,'html.parser')
    Tv_shows = soup.find('tbody',class_='lister-list').find_all('tr')

    for shows in Tv_shows:

        name = shows.find('td',class_='titleColumn').a.text
        rank = shows.find('td',class_='titleColumn').get_text(strip=True).split('.')[0]
        year_of_release = shows.find('td',class_='titleColumn').span.text.strip('()')
        rating = shows.find('td',class_='ratingColumn imdbRating').strong.text

        print(rank,name,year_of_release,rating)
        sheet.append([rank,name,year_of_release,rating])
except Exception as e:
    print(e)
```

```python
file.save('showsrating.csv')
```
